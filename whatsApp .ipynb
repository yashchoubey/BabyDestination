```python
text_file = open("WhatsApp Chat with Baby Destination SouthInd.txt", "r")
lines = text_file.readlines()
```

```python
lines[1]
```

```python
import unicodedata,unidecode,re
sentence_list=[]
for chat in lines:
    nfkd_form = unicodedata.normalize('NFKD', chat.decode('utf-8'))
    chat= unidecode.unidecode(u"".join([c for c in nfkd_form if not unicodedata.combining(c)]))
    
    if chat.count(':')>1:
        chat =chat.split(":")[2:]
        
    if 'Moms Event Sonakshi added' not in chat:
        chat=''.join(chat).replace("\n",' ')
        chat= re.sub(r'http\S+', '', chat)
        sentence_list.append(chat)
print sentence_list
```

```python
all_text=''.join(sentence_list)
all_text=all_text.lower()
```
