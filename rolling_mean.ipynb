```python
import pandas as pd
df = pd.read_excel('Data Analyst Test.xlsx', sheet_name=1,header=1,shuffle=False)
df['Date'] =pd.to_datetime(df.Date)
df=df.sort_values(by='Date').reset_index()
df['rm']=0
# df.drop(['Date'], axis=1, inplace=True)
df.drop(['Engaged users','reach','impressions'], axis=1, inplace=True)
print df
```

```python
df['rm'][0]=df['Likes'][0]
for i in range(1,len(df['rm'])):
    k=i-1
    sum=df['Likes'][i]    
    while k>=0 and (df['Date'][i]-df['Date'][k]).days<31:
        sum=sum+df['Likes'][k]
        k=k-1
    df['rm'][i]=sum/(i-k)
```
